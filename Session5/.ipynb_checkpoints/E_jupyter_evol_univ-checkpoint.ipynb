```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Evolution of the Universe with Planck 2015 cosmology (1)
from nbodykit.lab import cosmology
import astropy.units as u
import numpy as np

cosmo = cosmology.Planck15
z = np.logspace(-6,6,100)
tGyr = cosmo.time(z) * u.Gyr
t = tGyr.to(u.yr).value      # in year

#TODO
rho_r = cosmo.rho_r(z)
rho_m = cosmo.rho_m(z)
rho_l = cosmo.rho_lambda(z)

plt.plot(t, rho_r, label='radiation')
plt.plot(t, rho_m, label='matter')
plt.plot(t, rho_l, label=r'$\Lambda$')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Age of universe [yr]')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Amount of Dark Energy must increase for the density to stay constant!

```python
# Evolution of the Universe with Planck 2015 cosmology (2)

#Take the difference of the logarithms to avoid numerical issues with very large numbers
delta_rm = np.abs(np.log(rho_r / rho_m))
delta_ml = np.abs(np.log(rho_m / rho_l))


t1 = t[delta_rm.argmin()]
t2 = t[delta_ml.argmin()]

print('Radiation - matter transition: ' \
      '{0:.2g} yr'.format(t1))
print('Matter - dark energy transition: ' \
      '{0:.2g} yr'.format(t2))
```

```python
# Evolution of the Universe with Planck 2015 cosmology (3)
from scipy import stats

a = 1 / (1 + z)
logt = np.log(t)
loga = np.log(a)

range1 = (t > 10) & (t < 1e3)
range2 = (t > 1e6) & (t < 1e8)

# Linear fitting of the power indices
slp1, itc1, rval, pval, std = \
      stats.linregress(logt[range1], loga[range1])
slp2, itc2, rval, pval, std = \
      stats.linregress(logt[range2], loga[range2])

fit1 = np.exp(slp1 * logt + itc1)
fit2 = np.exp(slp2 * logt + itc2)

plt.plot(t, a) #Actual evolution
plt.plot(t, fit1, 'k:') #Fit of the first part
plt.plot(t, fit2, 'k:') #Fit of the second part

plt.annotate(r'$a \propto t^{{{0:.3g}}}$'.format(slp1), \
      (0.75,0.6), xycoords='axes fraction', fontsize=16)
plt.annotate(r'$a \propto t^{{{0:.3g}}}$'.format(slp2), \
      (0.20,0.1), xycoords='axes fraction', fontsize=16)

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Age of universe [yr]')
plt.ylabel('Scale factor')
plt.show()
```

The kink happens where we go from a radiation to a matter dominated Universe. The Friedmann equations change and so does the scale factor dependence on time.

```python
# Evolution of the Universe with Planck 2015 cosmology (4)
a = 1. / (1. + z)

rho_r = cosmo.rho_r(z)
rho_m = cosmo.rho_m(z)
rho_l = cosmo.rho_lambda(z)
rho_c = cosmo.rho_crit(z)

Or = rho_r / rho_c
Om = rho_m / rho_c
Ol = rho_l / rho_c

plt.plot(a, Or, label='radiation')
plt.plot(a, Om, label='matter')
plt.plot(a, Ol, label=r'$\Lambda$')

plt.xscale('log')
plt.yscale('log')
plt.xlim([1e-6,1])
plt.ylim([1e-5,1])
plt.xlabel('Scale factor')
plt.ylabel('Density parameter')
plt.legend()
plt.show()
```

```python
plt.plot(t, Or, label='radiation')
plt.plot(t, Om, label='matter')
plt.plot(t, Ol, label=r'$\Lambda$')

plt.xscale('log')
plt.yscale('log')
#plt.xlim([1e-6,1])
plt.ylim([1e-5,1])
plt.xlabel('Scale factor')
plt.ylabel('Density parameter')
plt.legend()
plt.show()
```

During the matter domination phase, the matter density is clearly dominating (nearly 1). Same for the other epochs. Note that because the critical density depends on time, the Lambda curve now also depends on time. We can check that three curves all sum up to 1 at every point.
