```python
# Evolution of Hubble parameter
from nbodykit.lab import cosmology
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#TODO
cosmo = cosmology.Planck15

amax = 100
cosmo = cosmo.clone(a_max=amax)
z = np.linspace(5, 1/amax - 1)

H = cosmo.efunc(z) * cosmo.h * 100
plt.plot(z, H)
plt.xlabel(r'$z$')
plt.ylabel(r'$H(z)\,[{\rm km}\,{\rm s}^{-1}\,' \
        r'{\rm Mpc}^{-1}]$')
plt.show()
```

```python
plt.plot(z, H/(1+z))
plt.yscale('log')
plt.xlabel(r'$z$')
plt.ylabel(r'$H(z) / (1+z) ' \
        r'\,[{\rm km}\,{\rm s}^{-1}\,{\rm Mpc}^{-1}]$')
plt.show()
```

We see that the expansion of the universe was decelerating, but in the future it will accelerate very much ($\frac{H}{1+z} = H * a = \dot a$). We can see this if we look at the evolution of the evolution of the density parameters.

```python
# Evolution of components of the Universe
rho_m = cosmo.rho_m(z)
rho_l = cosmo.rho_lambda(z)
rho_c = cosmo.rho_crit(z)

Om = rho_m / rho_c
OL = rho_l / rho_c

plt.plot(z, Om, label=r'$\Omega_{\rm m}$')
plt.plot(z, OL, label=r'$\Omega_\Lambda$')

plt.xlabel(r'$z$')
plt.ylabel(r'$\Omega$')
plt.legend()
plt.show()
```

In the past, the matter dominated. As a consequence, gravity played an important role and slowed down the expansion of the universe (note that the universe was always expanding). But recently, the DM/Lambda took over and therefore the expansion of the universe is accelerating again.
