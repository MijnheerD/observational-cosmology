```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

ngalaxies=5468756
L=2500 # Mpc h^-1
r = np.array([2.5, 7.5, 12.5, 17.5, 22.5, 27.5, 32.5, 37.5, 42.5, 47.5, 52.5, 57.5, \
            62.5, 67.5, 72.5, 77.5, 82.5, 87.5, 92.5, 97.5, 102.5, 107.5, 112.5, \
            117.5, 122.5, 127.5, 132.5, 137.5, 142.5, 147.5, 152.5, 157.5, 162.5, \
            167.5, 172.5, 177.5, 182.5, 187.5, 192.5, 197.5]) # Mpc h^-1

r1 = r - 2.5 # Mpc h^-1
r2 = r + 2.5 # Mpc h^-1
pairs = np.array([3.21395100e+06, 7.98631100e+06, 1.46335310e+07, 2.37986690e+07,
               3.56831090e+07, 5.04270100e+07, 6.80751500e+07, 8.87084200e+07,
               1.12279915e+08, 1.38942310e+08, 1.68618295e+08, 2.01354480e+08,
               2.37106790e+08, 2.75858730e+08, 3.17702780e+08, 3.62551625e+08,
               4.10493230e+08, 4.61729385e+08, 5.16313750e+08, 5.73975800e+08,
               6.34274350e+08, 6.97048250e+08, 7.62571100e+08, 8.30891400e+08,
               9.02348300e+08, 9.77014350e+08, 1.05493975e+09, 1.13605675e+09,
               1.22016245e+09, 1.30739095e+09, 1.39769280e+09, 1.49092370e+09,
               1.58709995e+09, 1.68628165e+09, 1.78838955e+09, 1.89383640e+09,
               2.00215300e+09, 2.11330340e+09, 2.22744280e+09, 2.34461935e+09])
```

```python
#TODO
npairs = ngalaxies*(ngalaxies-1)/2.

Nobs = pairs/npairs

plt.plot(r, Nobs, color='b', ls='-' , marker='o', label=r'$N_{\rm obs}$')
plt.xlabel(r'$s\,[\mathrm{h}^{-1}\,\mathrm{Mpc}]$')
plt.ylabel(r'N')
plt.legend()
plt.show()
```

```python
#TODO

Nrandom = (4 * np.pi * (r2**3 - r1**3) / 3)/L**3

#TODO
plt.plot(r, Nobs , color='b', ls='-' , marker='o', label=r'$N_{\rm obs}$')
plt.plot(r, Nrandom , color='r', ls='--', marker='*', label=r'$N_{\rm random}$')
plt.xlabel(r'$s\,[\mathrm{h}^{-1}\,\mathrm{Mpc}]$')
plt.ylabel(r'$N$')
plt.legend()
plt.show()
```

```python
#TODO
xi = Nobs/Nrandom - 1

plt.plot(r, r**2*xi , color='k', marker='*') #put r**2 to scale up the plot for more detail
plt.xlabel(r'$s\,[\mathrm{h}^{-1}\,\mathrm{Mpc}]$')
plt.ylabel(r'$s^2 \xi(s)$')
plt.show()
```

Looking at the plot above, we see that at small s the correlation function is positive, which means that the galaxies are clustered (more galaxies than expected from random distribution). At large s we see less galaxies, which confirms our reasoning even more. The bump we see at s = 100 is known as the _baryonic acoustic oscillation_.
